```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/compare_theta/theta_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.violinplot(x='Type', y='theta', data=df);
plt.title("Violin Plot, multiple groups");
```

```python
from scipy.stats import f_oneway

income_groups = [df.loc[df['Type']==type, 'theta'].values for type in df['Type'].dropna().unique()]
stat, p_value = f_oneway(*income_groups)
print(f"F Test: statistic={stat:.4f}, p-value={p_value:.4f}")
```
