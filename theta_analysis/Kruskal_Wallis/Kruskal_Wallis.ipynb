## Introduction

To test if scores on an ordinal scale are different across various categories we could use a so-called Kruskal-Wallis H test (Kruskal & Wallis, 1952). 

This will look at so-called rankings and not simply the median of each category. If you want to test truly if the median for each category is different, you should use Mood's Median Test (Mood, 1950).
